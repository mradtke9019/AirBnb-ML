```python
from afinn import Afinn
import pandas as pd
```

```python
afn = Afinn(language='en')
```

```python
df = pd.read_csv("./scrubbed-reviews.csv")
```

```python
df.head()
```

```python
for i, x in enumerate(df):
    print(x)
#     if i >= 5:
#         break
```

```python
# this will store the key value pair of the sentiments for a listing
# example: listing 123: sentiment value +55, or listing 13: sentiment value -1
listingSentiment = {}
listingSentimentCount = {}
```

```python
#  listing_id,id,reviewer_id,comments
for index, row in df.iterrows():
    try:
        listingId = int(row['listing_id'])
        if listingId not in listingSentiment:
            listingSentiment[listingId] = 0
            listingSentimentCount[listingId] = 0
    #     print(index,listingId, row['comments'])
        listingSentiment[listingId] += afn.score(row['comments'])
        listingSentimentCount[listingId] += 1
    except:
        print(index,listingId, row['comments'])
```

```python
print(len(listingSentiment))
print(len(listingSentimentCount))
```

```python
csvString = "id,sentiment,sentiment_average\n"
for lId in listingSentiment:
    average = 0
    if listingSentimentCount[lId] != 0:
        average = listingSentiment[lId]/listingSentimentCount[lId]
    csvString += str(lId) +"," + str(listingSentiment[lId]) + ","+ str(average) + "\n"
```

```python
file = open("sentiments.csv", "w")
n = file.write(csvString)
file.close()
```

```python
listingsDf = pd.read_csv("./scrubbed-listings.csv")
```

```python
header = "id,name,description,neighborhood_overview,host_id,host_duration_years,host_location,host_about,host_response_time,host_response_rate,host_acceptance_rate,host_is_superhost,host_neighbourhood,host_listings_count,host_total_listings_count,host_phone_verified,host_email_verified,host_work_email_verified,host_has_profile_pic,host_identity_verified,neighbourhood,neighbourhood_cleansed,property_type,room_type,accommodates,bathrooms,bedrooms,beds,amenities,price,minimum_nights,maximum_nights,minimum_minimum_nights,maximum_minimum_nights,minimum_maximum_nights,maximum_maximum_nights,minimum_nights_avg_ntm,maximum_nights_avg_ntm,has_availability,availability_30,availability_60,availability_90,availability_365,number_of_reviews,number_of_reviews_ltm,number_of_reviews_l30d,years_since_first_review,years_since_last_review,review_scores_rating,review_scores_accuracy,review_scores_cleanliness,review_scores_checkin,review_scores_communication,review_scores_location,review_scores_value,instant_bookable,calculated_host_listings_count,calculated_host_listings_count_entire_homes,calculated_host_listings_count_private_rooms,calculated_host_listings_count_shared_rooms,reviews_per_month"
headers = header.split(",")
sentimentFields = ["neighborhood_overview", "description", "host_about", "name"]
```

```python

for index, row in listingsDf.iterrows():
    for h in sentimentFields:
        cellValue = row[h]
        try:
#                print(h, row[h])
            cellValue = afn.score(cellValue)
        except:
            cellValue = 0
            print(index, h, row[h], cellValue)
        listingsDf.at[index,h]=cellValue
```

```python

listingsDf['review_sentiment'] = 0
listingsDf['sentiment_average'] = 0
for index, row in listingsDf.iterrows():
    ID = row['id']
    if ID in listingSentiment:
        if listingSentimentCount[ID] == 0:
            listingsDf.at[index,'review_sentiment']=0
            listingsDf.at[index,'sentiment_average']=0
        else:
            listingsDf.at[index,'review_sentiment']=listingSentiment[ID]
            listingsDf.at[index,'sentiment_average']=listingSentiment[ID]/listingSentimentCount[ID]
```

```python
listingsDf.to_csv("scrubbed-data.csv",index=False)

```
